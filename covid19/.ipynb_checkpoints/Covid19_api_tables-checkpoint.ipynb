```python
#TABLE 5 
import pandas as pd
import json
import requests

url = "https://api.covid19api.com/summary"
page = requests.get(url).json()
t = page['Global']
df=pd.DataFrame.from_dict(t, orient='index')
df.to_csv('Global summary',index=True,sep=',')
```
